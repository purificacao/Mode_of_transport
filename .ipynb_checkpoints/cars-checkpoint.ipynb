## One-way analysis of variance (ANOVA) for the 3 groups.

ANOVA is a Parametric Statistical Hypothesis Test. It tests whether the means of two or more independent groups are significantly different.

**There are three primary assumptions to apply this statistical 
test:**

   1. The responses for each group have a normal population distribution.
   2. These distributions have the same variance.
    - Homogeneity of variances
    - Homoscedasticity
   3. The data are independent.

**Interpretation**


   1. H0 (Null hypotheses): the means of the samples are equal.
   2. Ha ( Alternative hypotheses): one or more of the means of the samples are unequal.
   
Side note: Violations to the first two that are not extreme can be considered not serious. The sampling distribution of the test statistic is fairly robust, especially as sample size increases and more so if the sample sizes for all groups are equal.

```python
df_anova = df[['Salary','Transport']]
grps = pd.unique(df_anova.Transport.values)
d_data = {grp:df_anova['Salary'][df_anova.Transport == grp] for grp in grps}
```

```python
df_anova.assign(index = df_anova.groupby('Transport').cumcount()).pivot('index','Transport','Salary').plot(
    kind='box', figsize=(10,5))
plt.ylabel('Salary')
plt.xlabel('Transport')
plt.savefig('images/3.png', bbox_inches='tight', dpi=300);
```

```python
F, p = ss.f_oneway(d_data['2Wheeler'], d_data['Car'], d_data['Public Transport'])

print("p-value for significance is: ", p)

if p < 0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

**Another way to perform ANOVA is as follows:**

```python
model = ols(formula='Salary ~ C(Transport)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

From the above result, we can say that one or more of the means for the salary among the groups are statistically different at a significant level of 5%. Let's see whether or not the means of 2Wheeler and Public Transport groups are significantly different:

From ANOVA analysis itelf, we know that the average salary of employees who use different modes of transport are statistically significant, but ANOVA does not tell which modes of transports are significantly different from each other. To know the pairs of significant different modes of transports, we will perform multiple pairwise comparison (Post-hoc comparison) analysis using Tukey HSD test.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=df_anova.Salary,   # Data
                          groups=df_anova.Transport,  # Groups
                          alpha=0.05)  # Significance level

tukey.plot_simultaneous()  # Plot group confidence intervals

tukey.summary()     
```

From the above result, we conclude that the average salary of people who go by 2Wheeler or Public Transport are not significantly different, but the average salary of individuos who go by car is significantly different of individuals who go by 2Wheeler or Public Transport, at a 5% significant level!

### Now, let's if some of the ANOVA assumptions are violated

The Shapiro-Wilk test can be used to check whether or not the residuals are normally distributed. 

H0 : distribution of residuals = normal distribution\
Ha : distribution of residuals $\neq$ normal distribution

```python
w, pvalue = ss.shapiro(model.resid)
print(w, pvalue)
```

```python
alpha = 0.05

if pvalue > alpha:
    print('At a significance level of {}, the Shapiro-Wilk test suggests that the data is drawn from a normal distribution (fail to reject H0).'.format(alpha))
else:
    print('At a significance level of {}, the Shapiro-Wilk test suggests that the data is NOT drawn from a normal distribution (fail to reject H0).'.format(alpha))
```

### Checking the homogeneity of the variances

Since the residuals are not normally distributed, we can use the Levene test to check the homogeneity of variances (otherwise, we could use the Bartlett’s test):

```python
l, pvalue = ss.levene(d_data['2Wheeler'], d_data['Car'], d_data['Public Transport'] )
print(l, pvalue)
```

```python
alpha = 0.05

if pvalue > alpha:
    print('At a significance level of {}, the Levene test suggests that the variances are the same (H0 cannot be rejected).'.format(alpha))
else:
    print('At a significance level of {}, the Levene test suggests that the variances are NOT the same (H0 is rejected)'.format(alpha))
```

**We saw that the first two assumptions have been violeted (data drawn from a normal distribution and equal variances). Since these are the lesser important ones, we can still consider the insights taken from the ANOVA conducted.**

In order to get a more robust result, we should've used the same sample size for all three groups, so we could, for example, undersample **Public Transport** and **2Wheeler** to 35, so that they would match with the sample size of car users.

## Let's see the association strenght between the categorical features (Engineer, MBA, license, Gen and Transport )

For that, we're goint to be using the **[Theil'U](https://en.wikipedia.org/wiki/Uncertainty_coefficient)**

```python
def conditional_entropy(x,y):
    #x entropy, given y
    y_counter = Counter(y)
    xy_counter = Counter(list(zip(x,y)))
    total_occurrences = sum(y_counter.values())
    entropy = 0
    for xy in xy_counter.keys():
        p_xy = xy_counter[xy] / total_occurrences
        p_y = y_counter[xy[1]] / total_occurrences
        entropy += p_xy * math.log(p_y/p_xy)
    return entropy

def theil_u(x,y):
    s_xy = conditional_entropy(x,y)
    x_counter = Counter(x)
    total_occurrences = sum(x_counter.values())
    p_x = list(map(lambda n: n/total_occurrences, x_counter.values()))
    s_x = ss.entropy(p_x)
    if s_x == 0:
        return 1
    else:
        return (s_x - s_xy) / s_x
```

```python
c = []
for a in category.columns:
    for b in category.columns:
        c.append(theil_u(category[a], category[b]))
        
d = np.array(c).reshape(int(np.sqrt(len(c))),int(np.sqrt(len(c))))
corr = pd.DataFrame(data=d, columns=category.columns, index=category.columns)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=ax, fmt='.3f')
plt.setp(ax.get_yticklabels(), ha="right", rotation=0)
plt.setp(ax.get_xticklabels(), ha="right", rotation=45)

plt.title('Heatmap showing the strenght of association between each pair of category', 
fontdict={'fontsize': 13}, pad=10)
plt.savefig('images/9.png', bbox_inches='tight', dpi=300);
```

# Logistic model

How much of $Y$ can we explain, given $X$: $\hspace{.5cm}$ ${\displaystyle p_{i}=\operatorname {E} \left(\left.{\frac {Y_{i}}{n_{i}}}\right|X_{i}\right)\,\!}$


${\displaystyle \operatorname {logit} (p_{i})=\ln \left({\frac {p_{i}}{1-p_{i}}}\right)=\beta _{0}+\beta _{1}x_{1,i}+\cdots +\beta _{k}x_{k,i}.} {\displaystyle \operatorname {logit} (p_{i})=\ln \left({\frac {p_{i}}{1-p_{i}}}\right)=\beta _{0}+\beta _{1}x_{1,i}+\cdots +\beta _{k}x_{k,i}.}$


${\displaystyle p_{i}={\frac {1}{1+e^{-(\beta _{0}+\beta _{1}x_{1,i}+\cdots +\beta _{k}x_{k,i})}}}}$

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model = LogisticRegression(penalty='l2', solver='saga', class_weight='balanced')
result = model.fit(X_train, y_train)
```

```python
from math import exp
prediction_test = model.predict(X_test)
# Print the prediction accuracy
print('Acuracy: %.2f \n' %(metrics.accuracy_score(y_test, prediction_test)))
```

```python
weights = pd.Series(model.coef_[0], index=X.columns.values)
weights.sort_values(ascending = False).plot(kind='bar');
```

```python
def plotPrecisionRecallThreshold(y_test, pred_prob):
    precision, recall, thresholds = metrics.precision_recall_curve(y_test, pred_prob) 
    pr_auc = metrics.auc(recall, precision)
    plt.figure(figsize=(8,5))
    plt.title("Precisão-Recall vs Threshold")
    plt.plot(thresholds, precision[: -1], "r--", label="Precision")
    plt.plot(thresholds, recall[: -1], "k--", label="Recall")
    plt.ylabel("Precision, Recall")
    plt.xlabel("Threshold")
    plt.legend(loc="lower left")
    plt.ylim([0,1])
```

```python
predicted_proba=model.predict_proba(X_test)
plotPrecisionRecallThreshold(y_test, predicted_proba[:,1])
```

```python
def areaUnderROC(y_test, pred_proba):
    precision, recall, threshold = metrics.precision_recall_curve(y_test, pred_proba)
    return metrics.auc(recall, precision)
```

```python
def plotROC(y_test,pred_prob):
    fpr, tpr, threshold= metrics.roc_curve(y_test, pred_prob)
    plt.figure(figsize=(8,5))
    plt.title("ROC curve")
    sns.lineplot(x=fpr,y=tpr, color="red")
    plt.ylabel("True Positive Rate")
    plt.xlabel("False Positive Rate")
```

```python
plotROC(y_test, predicted_proba[:,1])
print('Area Under ROC: %.2f' % areaUnderROC(y_test, predicted_proba[:,1]))
```
